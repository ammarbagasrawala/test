```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread("/content/inside_out.webp",0)
```

```python
cv2_imshow(image)
```

```python
low_sigma = cv2.GaussianBlur(image,(3,3),0)
high_sigma = cv2.GaussianBlur(image,(5,5),0)

DOG_different_kernel= low_sigma - high_sigma

cv2_imshow(DOG_different_kernel)
```

```python
high_sigma = cv2.GaussianBlur(image,(5,5),21)
low_sigma = cv2.GaussianBlur(image,(5,5),3)

DOG_different_sigma = low_sigma - high_sigma

cv2_imshow(DOG_different_sigma )
```

```python
x = cv2.GaussianBlur(image,(3,3),21)
y = cv2.GaussianBlur(image,(5,5),3)

DOG = x-y

cv2_imshow(DOG)
```
