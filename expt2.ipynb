```python
# Image Transformations
# 1. Image Translation

# OpenCV provides two transformation functions, cv2.warpAffine and cv2.warpPerspective, with which you can have all kinds of transformations. 
# cv2.warpAffine takes a 2x3 transformation matrix while cv2.warpPerspective takes a 3x3 transformation matrix as input.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title('Original Image')
plt.imshow(img)
```

```python
rows, cols, channels = img.shape
trans_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
trans_img = cv2.warpAffine(img, trans_matrix, (cols, rows))

plt.title(f'Translated image, Sx={trans_matrix[0][2]}, Sy={trans_matrix[1][2]}')
plt.imshow(trans_img)
```

```python
transform_mat = np.float32([[-1,  0, cols],
                [0, 1, 0],
                [0,  0, 1]])
reflected_img = cv2.warpPerspective(img, transform_mat,(cols,rows))
plt.title('Reflected along x')
plt.imshow(reflected_img)
```

```python
transform_mat = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])
reflected_img = cv2.warpPerspective(img, transform_mat,(cols,rows))
plt.title('Reflected along y')
plt.imshow(reflected_img)
```

```python
angle, scale = 45,0.6
rotated_img = cv2.warpAffine(img, cv2.getRotationMatrix2D((cols//2, rows//2), angle, scale), (cols, rows))
plt.title(f'Rotated image Angle:{angle} Shrinkage:{1-scale}')
plt.imshow(rotated_img)
```

```python
scale_mat = np.float32([[2, 0, 0], [0, 4, 0]])
scale_img = cv2.warpAffine(img, scale_mat, (cols, rows))
plt.title(f'Scaled image sx:{scale_mat[0][0]} sy:{scale_mat[1][1]}')
plt.imshow(scale_img)
```

```python
start_row, end_row = 0, 100
start_col, end_col = 50, 100
cropped_img = img[start_row:end_row, start_col:end_col, :]
plt.title(f'Cropped Image')
plt.imshow(cropped_img)
```

```python
shearing_x = np.float32([[1, 0.5, 0], [0, 1, 0]])
shearx_img = cv2.warpAffine(img, shearing_x, (cols, rows))
plt.title(f'Shearing in x axis Shx:{shearing_x[0][1]}')
plt.imshow(shearx_img)
```

```python
shearing = np.float32([[1, 0.5, 0], [0.6, 1, 0]])
shear_img = cv2.warpAffine(img, shearing,(cols, rows))
plt.title(f'Shearing in x,y axis Shx:{shearing_x[0][1]} Shy:{shearing_x[1][0]}')
plt.imshow(shear_img)
```
