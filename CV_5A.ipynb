```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread("/content/kungfu_panda.jpg")
```

```python
cv2_imshow(image)
```

SOBEL FILTER

```python
F_sobel= np.float32([[2,2,0],
             [2,0,-2],
             [0,-2,-2]])

F_combined = cv2.filter2D(image, -1, F_sobel)

cv2_imshow(F_combined)
```

LAPLACIAN

```python
laplacian_image = cv2.Laplacian(image,ksize=3,ddepth=-1)
cv2_imshow(laplacian_image)
```

```python
gaussian_image = cv2.GaussianBlur(image,ksize=(3,3),sigmaX=5,sigmaY=5)
laplacian_of_gaussian_image = cv2.Laplacian(gaussian_image,ksize=3,ddepth=-1)
cv2_imshow(laplacian_of_gaussian_image)
```

```python
canny_image = cv2.Canny(image, 0, 255)
cv2_imshow(canny_image)
```
